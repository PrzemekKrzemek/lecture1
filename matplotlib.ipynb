```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
data = pd.read_csv("/content/lecture1/Salary Data (2).csv")
print(data)
```

```python
fig, ax = plt.subplots()
x = data["Years of Experience"]
y = data["Salary"]
ax.scatter(x,y)
ax.set_xlabel("Years of Experience")
ax.set_ylabel("Salary")
ax.set_title("Salary vs Experience")
```

```python
fig, ax = plt.subplots()
ax.scatter(x, y, color='blue', alpha=0.6, edgecolors='black')

m, b = np.polyfit(x, y, 1)
ax.plot(x, m*x + b, color='red', label="Trend Line")

ax.set_xlabel("Years of Experience")
ax.set_ylabel("Salary")
ax.set_title("Salary vs Experience")
ax.legend()
plt.show()
```
